## Viewing changes in energy sources over time

```python
fossil_fuels = np.concatenate((arange(19,31), arange(63,75), arange(141,146)))
electricity = np.arange(127,141)
```

```python
Zs = []
xs = []
Ys = []
Fs = []
F_Ys = []
D_cbas = []
D_pbas = []

for year in years:
    Z = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/Z.csv', header = 0, index_col = 0)
    x = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/x.csv', index_col = 0)
    Y = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/Y.csv', header = 0, index_col = 0)
    F = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/F.csv', header = 0, index_col = 0)
    F_Y = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/F_Y.csv', header = 0, index_col = 0)
    D_cba = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/D_cba.csv', header = 0, index_col = 0)
    D_pba = read_csv(path_to_data + 'IOT_' + str(year) + '_pxp/D_pba.csv', header = 0, index_col = 0)

    Zs.append(Z)
    xs.append(x)
    Ys.append(Y)
    Fs.append(F)
    F_Ys.append(F_Y)
    D_cbas.append(D_cba)
    D_pbas.append(D_pba)
```
